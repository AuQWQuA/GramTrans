```python
import json
with open('newgrammarmbpp.json', "r") as f:
    data = json.load(f)
with open("grammar.json", "r") as f:
    old_data = json.load(f)
```

```python
old_data['rules'].keys()
```

```python
data['rules'].keys()
```

```python
newrules = {}
for rule in old_data['rules'].keys():
    if rule in data['rules'].keys():
        newrules[rule] = data['rules'][rule]
        
```

```python
data['rules'] = newrules
```

```python
with open('newgrammarmbpp.json', "w") as f:
    json.dump(data, f, indent=2)
```
